```python
import numpy as np
import seaborn as sns
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
```

Define the 'tips' data set within the Seaborn library

```python
df = sns.load_dataset("tips")
df.head()
```

Find the sum, min, max and average of the total_bill values by "Time" variable.

```python
df.groupby("time", observed=False).agg({"total_bill": ["sum", "min", "max", "mean"]})
```

Find the sum, min, max and average of total_bill values by 'days' and 'time'

```python
df.groupby(["day", "time"], observed=False).agg({"total_bill": ["sum", "min", "max", "mean"]})
```

Find the sum, min, max and average of the total_bill by 'lunch time' and 'female' customers according to day

```python
filtered_df = df[(df["time"] == "Lunch") & (df["sex"] == "Female")]
result = filtered_df.groupby("day", observed=False).agg({"total_bill": ["sum", "min", "max", "mean"],"tip": ["sum", "min", "max", "mean"]})
print(result)
```

Average of orders with 'size' less than 3 and 'total_bill' greater than 10?

```python
order_avg_df = df[(df["size"] < 3) & (df["total_bill"] > 10)]
order_avg_mean = order_avg_df["total_bill"].mean()
print(order_avg_mean)
```

Create a new variable called total_bill_tip_sum. It will return the totalbill and tip paid by each customer

```python
df["total_bill_tip_sum"] = df.apply(lambda x: x["total_bill"] + x["tip"], axis=1)
df.head()
```

Sort from largest to smallest by the total_bill_tip_sum variable and assign the first 30 people to a new dataframe.

```python
new_df = df.sort_values(by="total_bill_tip_sum", ascending=False).head(30)
print(new_df)
```
